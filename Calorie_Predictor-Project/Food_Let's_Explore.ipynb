```python
import numpy as np
import pandas as pd 
```

```python
food_df = pd.read_csv('Food_Data.csv')
food_df.head()
```

```python
food_df.Food
```

```python
food_df.Calcium = food_df.Calcium.str.replace('MG', '')
food_df.Iron = food_df.Iron.str.replace('MG', '')
food_df.Sodium = food_df.Sodium.str.replace('MG', '')
food_df['Vitamin A'] = food_df['Vitamin A'].str.replace('MG', '')
food_df['Vitamin C'] = food_df['Vitamin C'].str.replace('MG', '')
food_df.Cholestrol = food_df.Cholestrol.str.replace('MG', '')
food_df.Protein = food_df.Protein.str.replace('G', '')
food_df.Carbs = food_df.Carbs.str.replace('G', '')
food_df.Fats = food_df.Fats.str.replace('G', '')
food_df.Calories = food_df.Calories.str.replace('KCAL', '')
food_df.Sugars = food_df.Sugars.str.replace('G', '')
food_df
```

```python
food_df = food_df.iloc[:, 1:].astype('float')
food_df.info()
```

```python
import seaborn as sns
sns.heatmap(food_df.corr(), annot=True)
```

```python
food_df.isnull().value_counts()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy = 'mean', missing_values = np.nan)
imp = imp.fit(food_df[['Calcium']])
food_df['Calcium'] = imp.transform(food_df[['Calcium']])
imp = imp.fit(food_df[['Iron']])
food_df['Iron'] = imp.transform(food_df[['Iron']])
imp = imp.fit(food_df[['Sodium']])
food_df['Sodium'] = imp.transform(food_df[['Sodium']])
imp = imp.fit(food_df[['Vitamin A']])
food_df['Vitamin A'] = imp.transform(food_df[['Vitamin A']])
imp = imp.fit(food_df[['Vitamin C']])
food_df['Vitamin C'] = imp.transform(food_df[['Vitamin C']])
imp = imp.fit(food_df[['Cholestrol']])
food_df['Cholestrol'] = imp.transform(food_df[['Cholestrol']])
imp = imp.fit(food_df[['Calories']])
food_df['Calories'] = imp.transform(food_df[['Calories']])
imp = imp.fit(food_df[['Sugars']])
food_df['Sugars'] = imp.transform(food_df[['Sugars']])
food_df
```

```python
from sklearn.model_selection import train_test_split
X = food_df.drop('Calories', axis = 1)
y = food_df[['Calories']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, predictions)
mse
```

```python
mae = mean_absolute_error(y_test, predictions)
mae
```

```python
rmse = np.sqrt(mse)
rmse
```
